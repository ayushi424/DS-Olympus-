# Feture Engineering:

```python
df2=df.copy()
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.drop(['sales'],axis=1,inplace=True)
```

```python
df2.columns
```

```python
df2['salary'].unique()
```

```python
salary_dummies=pd.get_dummies(df2['salary'],drop_first=True)
```

```python
salary_dummies
```

```python
df2=pd.concat([df2,salary_dummies],axis=1)
```

```python
df2.head()
```

```python
df2.drop(['salary'],axis=1,inplace=True)
```

```python
df2.loc[308]
```

# Train-Test Split:

```python
x=df2.drop(labels='left',axis=1)
y=df2['left']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=14)
```

```python
len(X_train)
```

```python
len(X_test)
```

#  Hyper Parameter Tuning for model selection:

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
import pickle
```

```python
model_params={
    
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[14,28,9]
        }
        
    },
    
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[14,28,9]
        }
    }
        
    
}
```

```python
scores=[]
for model_name,mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(X_train,y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
df=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

### Model Implementation Using Rndom Forest Classifier:

```python
model=RandomForestClassifier(n_estimators=28,bootstrap=True,max_features='sqrt')
model.fit(X_train,y_train)
```

```python
#model.estimators_
```

```python
#plt.figure(figsize=(15,10))
#tree.plot_tree(model.estimators_[1],filled=True)
```

```python
plt.figure(figsize=(30,15))
for i in range(1):
    tree.plot_tree(model.estimators_[i],filled=True)
```

### Model Evaluation:
